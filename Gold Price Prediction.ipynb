```python
df = data[['Date', 'INR']].copy()
```

```python
print(type(df))
print(df.head()) 
```

```python
df['INR'] = df['INR'].astype(str).str.replace(",", "", regex=True)
df['INR'] = pd.to_numeric(df['INR'], errors='coerce')
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Date'])
print(df.head())
```

```python
df.dropna(subset=['INR'],inplace=True)
print(df.head())
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.set_index('Date',inplace=True)
```

```python
df.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df.index,df['INR'],label='Gold Price')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Gold Price Over Time')
plt.legend()
plt.show()
```

Null Hypothesis: Not Stationary
Alternate Hypothesis: Stationary

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(df['INR'])
```

```python
print(f'Test Statistics: {result[0]}')
print(f'p-value: {result[1]}')
print(f'Critical Value: {result[2]}')
print(f'No:of Lags: {result[3]}')

if result[1] > 0.05:
    print("Series is not Stationary")
else:
    print("Series is Stationary")
```
